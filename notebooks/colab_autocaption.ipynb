## 2. Start Server dengan Ngrok (Port Forwarding)

```python
# 1. Start server di background dulu
import os
import threading
import uvicorn

PORT = 8000
WORK_DIR = "/content/autocaption_web"
os.chdir(WORK_DIR)

def run_server():
    uvicorn.run("app.main:app", host="0.0.0.0", port=PORT)

thread = threading.Thread(target=run_server, daemon=True)
thread.start()

# Tunggu server siap
import time
time.sleep(3)

# 2. Port forwarding dengan Ngrok
# Dapatkan token gratis di https://ngrok.com (opsional, untuk session lebih lama)
from pyngrok import ngrok

# os.environ["NGROK_AUTH_TOKEN"] = "your_token"  # uncomment jika punya token

public_url = ngrok.connect(PORT)
print(f"\n🌐 Aplikasi tersedia di: {public_url}")
print(f"   Buka URL di browser untuk upload video/audio dan download subtitle.")
```

```python
# Server sudah berjalan di background. Jangan stop/restart runtime agar tetap bisa diakses.
# Untuk menghentikan: Runtime → Interrupt execution
print("Server berjalan. Gunakan URL ngrok di atas untuk mengakses aplikasi.")
```
